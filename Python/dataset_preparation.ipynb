```python
import pandas as pd
import random

df = pd.read_csv("Car_Sales_Reduced.csv")

df.head()
```

```python
df['SaleID'] = range(1, len(df) + 1)
```

```python
random.seed(42)
salesperson_df = df[['Salesperson']].drop_duplicates().reset_index(drop=True)

# Create SalespersonID
salesperson_df['SalespersonID'] = range(1, len(salesperson_df) + 1)

# Enrich data
regions = ['North', 'South', 'East', 'West']
experience = [1, 2, 3, 5, 10]

salesperson_df['Region'] = [random.choice(regions) for _ in range(len(salesperson_df))]
salesperson_df['Experience'] = [random.choice(experience) for _ in range(len(salesperson_df))]


salesperson_df = salesperson_df[['SalespersonID', 'Salesperson', 'Region', 'Experience']]

# Save
salesperson_df.to_excel("Salesperson.xlsx", index=False)
```

```python
random.seed(42)

car_df = df[['Car Make', 'Car Model', 'Car Year']].drop_duplicates().reset_index(drop=True)
car_df['CarID'] = range(1, len(car_df) + 1)

categories = ['SUV', 'Sedan', 'Hatchback', 'Luxury', 'Truck']
# These choices are also "locked" by the seed
car_df['Category'] = [random.choice(categories) for _ in range(len(car_df))]

car_df = car_df[['CarID', 'Car Make', 'Car Model', 'Car Year', 'Category']]
car_df.to_csv("Car.csv", index=False)
```

```python
df = df.merge(salesperson_df, on='Salesperson')
df = df.merge(car_df, on=['Car Make', 'Car Model', 'Car Year'])
```

```python
sales_df = df[['SaleID',
               'Date',
               'SalespersonID',
               'Customer Name',
               'CarID',
               'Sale Price',
               'Commission Rate',
               'Commission Earned']]

# Save
sales_df.to_csv("Sales.csv", index=False)
```
